### Contr√¥ler le nombre de chiffres affich√©s apr√®s la virgule

```python
valeur_experimentale = 34.7879783705980890
print(valeur_experimentale)
```

Vous l'avez sans doute appris en TP, il n'est pas pertinent d'afficher une valeur num√©rique avec tous les chiffres apr√®s la virgule.

L'√©criture format√©e permet de limiter le nombre de chiffres apr√®s la virgule.

Avec un seul chiffre apr√®s la virgule :

```python
print(f"{valeur_experimentale:.1f}")
```

üìù Remarques : 

- Le symbole `:` s√©pare le nom de la variable (`valeur_experimentale`) de l'op√©rateur de formatage (`.1f`).
- L'op√©rateur de formatage `.1f` signifie qu'on souhaite afficher un nombre r√©el (*float* -> `f`) avec un chiffre apr√®s la virgule (`.1`)
- C'est bien un arrondi qui est r√©alis√© ici.

Avec deux chiffres apr√®s la virgule :

```python
print(f"{valeur_experimentale:.2f}")
```

Avec trois chiffres apr√®s la virgule :

```python
print(f"{valeur_experimentale:.3f}")
```
