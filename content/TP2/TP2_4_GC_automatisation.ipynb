# Automatisation du calcul du pourcentage de GC et de la température de fusion



## Automatisation du calcul du %GC

L'objectif est de calculer le %GC de la séquence suivante :

`TAGATCCTTGGGTCCATTTGACGTAGTCTATGACG`

sans compter à la main le nombre de bases `A`, `T`, `C` et `G`, mais en automatisant ce comptage avec une boucle `for` et des tests `if` et `elif`.

On souhaite ensuite afficher la valeur du %GC avec un chiffre après la virgule, sous la forme :

```
Le pourcentage de GC vaut : xx.x
```

Exécutez la cellule suivante pour créer la variable qui contiendra la séquence à étudier :

````python
sequence = "TAGATCCTTGGGTCCATTTGACGTAGTCTATGACG"
````

Dans la cellule ci-dessous :

- **Comptez automatiquement** le nombre de bases avec une boucle `for` et des tests `if` et `elif`.
- Stockez le nombre de bases dans les variables `nb_A`, `nb_T`, `nb_C` et `nb_G`.
- Après la boucle, affichez le contenu de ces 4 variables et vérifiez avec vos voisin·e·s que vous obtenez des réponses identiques.

:::{tip}

N'hésitez pas à jeter un oeil au notebook `TP2_2_boucles_tests.ipynb`

:::

:::{admonition} Éléments de réponse
:class: tip
:class: dropdown

Voici un exemple de code possible avec :

```python
nb_A = 0
nb_T = 0
nb_C = 0
nb_G = 0

for base in sequence:
    if base == "A":
        nb_A = nb_A + 1
    elif base == "T":
        nb_T = nb_T + 1
    elif base == "C":
        nb_C = nb_C + 1
    elif base == "G":
        nb_G = nb_G + 1

print(nb_A, nb_T, nb_C, nb_G)
```

N'oubliez pas d'initialiser les variables `nb_A`, `nb_T`, `nb_C` et `nb_G` avant la boucle, c'est-à-dire de les créer avec la valeur 0.

Vous devriez obtenir les valeurs suivantes pour les variables `nb_A`, `nb_T`, `nb_C` et `nb_G` (dans le même ordre) :

```none
7 12 7 9
```
